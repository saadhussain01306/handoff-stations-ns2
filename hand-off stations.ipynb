```python
#OPI-RLI-HO (Optimization Parameter Interpolation-Received Level Indicator-Handover)
import csv
import matplotlib.pyplot as plt

# Read data from CSV file
data = []
with open('data_opi.csv', 'r') as file:
    reader = csv.reader(file)
    header = next(reader)
    for row in reader:
        data.append([float(val) for val in row])

# Extract columns
optimization_parameter = [row[0] for row in data]
received_level_indicator = [row[1] for row in data]
handover_decision = [row[2] for row in data]

# Plotting
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(optimization_parameter, bins=20, color='blue', alpha=0.7)
plt.xlabel('Optimization Parameter')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(received_level_indicator, bins=20, color='green', alpha=0.7)
plt.xlabel('Received Level Indicator')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(handover_decision, bins=[-0.5, 0.5, 1.5], color='red', alpha=0.7, rwidth=0.8)
plt.xticks([0, 1], ['No Handover', 'Handover'])
plt.xlabel('Handover Decision')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#Traditional Hand Over method
import csv
import matplotlib.pyplot as plt

# Read data from CSV file
data = []
with open('data_trad.csv', 'r') as file:
    reader = csv.reader(file)
    header = next(reader)
    for row in reader:
        data.append([float(val) for val in row])

# Extract columns
rss = [row[0] for row in data]
snr = [row[1] for row in data]
cell_load = [row[2] for row in data]
coverage_area = [row[3] for row in data]
mobility_speed = [row[4] for row in data]
handover_threshold = [row[5] for row in data]

# Plotting
plt.figure(figsize=(10, 6))

plt.subplot(2, 3, 1)
plt.hist(rss, bins=20, color='blue', alpha=0.7)
plt.xlabel('Received Signal Strength (RSS)')
plt.ylabel('Frequency')

plt.subplot(2, 3, 2)
plt.hist(snr, bins=20, color='green', alpha=0.7)
plt.xlabel('Signal-to-Noise Ratio (SNR)')
plt.ylabel('Frequency')

plt.subplot(2, 3, 3)
plt.hist(cell_load, bins=20, color='red', alpha=0.7)
plt.xlabel('Cell Load')
plt.ylabel('Frequency')

plt.subplot(2, 3, 4)
plt.hist(coverage_area, bins=20, color='orange', alpha=0.7)
plt.xlabel('Cell Coverage Area')
plt.ylabel('Frequency')

plt.subplot(2, 3, 5)
plt.hist(mobility_speed, bins=20, color='purple', alpha=0.7)
plt.xlabel('Mobility Speed')
plt.ylabel('Frequency')

plt.subplot(2, 3, 6)
plt.hist(handover_threshold, bins=20, color='brown', alpha=0.7)
plt.xlabel('Handover Threshold')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#Group Cell HandOver [GCHO] method 
import csv
import matplotlib.pyplot as plt

# Read data from CSV file
data = []
with open('data_group.csv', 'r') as file:
    reader = csv.reader(file)
    header = next(reader)
    for row in reader:
        data.append([float(val) for val in row])

# Extract columns
group_size = [row[0] for row in data]
interference_level = [row[1] for row in data]
group_load = [row[2] for row in data]
group_coverage_area = [row[3] for row in data]

# Plotting
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.hist(group_size, bins=20, color='blue', alpha=0.7)
plt.xlabel('Group Size')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
plt.hist(interference_level, bins=20, color='green', alpha=0.7)
plt.xlabel('Interference Level')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
plt.hist(group_load, bins=20, color='red', alpha=0.7)
plt.xlabel('Group Load')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.hist(group_coverage_area, bins=20, color='orange', alpha=0.7)
plt.xlabel('Group Coverage Area')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
